## Plotting

### Line Plot
menunjukkan perubahan pada waktu

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

# styling
mpl.style.use(['ggplot'])
```

```python
# plot garis untuk data imigran haiti

years = list(range(1980, 2014))
haiti = df_can.loc['Haiti', years]
haiti
```

```python
haiti.plot();

plt.title("Imigration from Haiti");
plt.ylabel("Jumlah Imigran");
plt.xlabel("Tahun");

plt.text(2000,6000, 'Gempa Haiti 2010');
```

**Pertanyaan 1** : Bagaimana perbandingan jumlah imigran China dan India dari Tahun 1980-2013 ke Canada?

```python
negara = ['China', 'India']
china_india = df_can.loc[negara,years]
china_india
```

```python
china_india.plot();
```

```python
# transpose

CI_transpose = china_india.T
CI_transpose.head()
```

```python
CI_transpose.plot();

plt.title("Imigran dari China dan India");
plt.ylabel("Jumlah Imigran");
plt.xlabel("Tahun");
```

**Pertanyaan 2** : Bandingkan tren dari 5 negara teratas yang menyumbang imingran ke Canada!

```python
top5_index = df_can['Total'].sort_values(ascending=False).head().index
top5_df = df_can.loc[top5_index, years]
top5_df = top5_df.T
```

```python
# figsize (panjang, lebar)

top5_df.plot(figsize=(15, 8));

plt.title("Imigran dari 5 Negara Teratas");
plt.ylabel("Jumlah Imigran");
plt.xlabel("Tahun");
```

## Area Plot

```python
top5_df.plot(kind = 'area',
            stacked = False,
            figsize=(15, 8));

plt.title("Imigran dari 5 Negara Teratas");
plt.ylabel("Jumlah Imigran");
plt.xlabel("Tahun");
```

## Histogram

untuk mereprentasikan data yang bersifat continu, sedangkan bar chart untuk data kategorik/diskrit.

```python
df_2013 = df_can[2013]
df_2013
```

```python
# numpy

count, bin_edges = np.histogram(df_2013)

print(count)
print(bin_edges)
```

```python
df_2013.plot(kind='hist',
            bins=10,
            xticks= bin_edges,
            figsize=(8, 5));

plt.title("Histogram dari Data Imigran Tahun 2013");
plt.ylabel("Jumlah Negara");
plt.xlabel("Jumlah Imigran");
```

**Pertanyaan 3** : Bagaimana distribusi dari Imigran Denmark, Norwegia, dan Swedia?

```python
df_dns = df_can.loc[['Denmark', 'Norway', 'Sweden'], years]
df_dns
```

```python
df_dns.T.plot(kind='hist',
                alpha=0.5,
                bins=15,
                figsize=(10,6),
                color=['coral', 'darkslateblue', 'mediumseagreen']);
```

## Pie Chart

```python
df_can
```

```python
df_continent = df_can.groupby('Continent').sum()
df_continent
```

```python
df_continent['Total'].plot(kind='pie',
                            figsize=(5,6),
                            autopct="%.2f%%",
                            startangle=90,
                            shadow=True);
```

```python
explode_list = [0.1,0,0,0,0.1,0.1]
df_continent['Total'].plot(kind='pie',
                            figsize=(15,6),         # edit ukuran chart
                            autopct="%.1f%%",       # menambahkan label persentase
                            startangle=90,          # rotate pie chart
                            shadow=True,            # menambahkan bayangan
                            labels=None,            # label
                            pctdistance=1.15,       # edit jarak label persentase
                            explode=explode_list);  # mengeluarkan bagian pie chart

plt.legend(labels=df_continent.index);
plt.title("Data Imigrasi sesuai Continent", y=1.12);
plt.axis('equal');      # edit luas latar belakang
```
